## SVM 模型（pytorch版）

```python
class LinearSVM(nn.Module):
    def __init__(self, input_size, output_size):
        super(LinearSVM, self).__init__()
        self.linear = nn.Linear(input_size,output_size)
        
    def forward(self, x):

        y_pred = self.linear(x)
        return y_pred
    
```

## 训练模型

```python
def train_svm_model(model, batch_data, batch_label,LEARNING_RATE= 0.0001, EPOCHS = 50):
    
    
    best_zhunque = 0
    best_zhaohui = 0
    best_jingque = 0

    optimizer = torch.optim.SGD(model.parameters(), lr=LEARNING_RATE)
    model.train()
    #loss_fn = nn.MSELoss(reduction = 'sum') #这里只是 loss function 
    if torch.cuda.is_available():
        ''' 将数据转换到GPU运算 '''
        model = model.cuda()
        batch_data = torch.Tensor(batch_data).cuda()
        batch_label = torch.Tensor(batch_label).cuda()

    for iterater in range(EPOCHS):
        TP, TN, FP, FN = 0.1, 0.1, 0.1, 0.1 # 防止分母为0
        print('*'*20,"第",iterater,"轮",'*'*20)
        
        for i in range(len(batch_data)):
        
            x = batch_data[i]
            y = batch_label[i]
            # Forward pass
            y_pred = model(x) # model 会直接 进行 forward pass  即model.forward()

            # Compute loss
            #loss = loss_fn(y_pred , y) # 这里是一个 compute graph
            loss = torch.mean(torch.clamp(1 - y_pred.t() * y, min=0))  # hinge loss
            loss += torch.mean(model.linear.weight ** 2)  # l2 penalty
           # print(y_pred,y)
            if i%200 ==0:
                print(i,loss.item())
            if y_pred<0.5 and y == 0 : TN += 1
            if y_pred<0.5 and y == 1 : FP += 1
            if y_pred>0.5 and y == 0 : FN += 1
            if y_pred>0.5 and y == 1 : TP += 1
            ''' 清空上一次计算的梯度 '''
            optimizer.zero_grad()
            # Backwrad pass
            loss.backward()
            optimizer.step() #自动更新优化 model 的参数

        
        print(TP, TN, FP, FN)
        a = (TP+TN)/(TP+TN+FN+FP)
        if a>best_zhunque:
            best_zhunque = a
            best_jingque = TP/(TP+FP)
            best_zhaohui = TP/(FN+TP)
        print("测试完毕， 准确率是：" + str((TP+TN)/(TP+TN+FN+FP)))
        print('精确率'+str(TP/(TP+FP)))
        print('召回率'+str(TP/(FN+TP)))
    print("最高准确率：",best_zhunque)
    print("最高精确率：",best_jingque)
    print("最高召回率：",best_zhaohui)
```
