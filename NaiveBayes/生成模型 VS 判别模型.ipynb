# 生成模型 VS 判别模型
### 从一个例子开始：

<br><br>现在我们有一个数据集${（X_1,Y_1）,（X_2,Y_2）,...,（X_n,Y_n）,X_i=(x_1,x_2,...,x_m),y_i={0 or 1}$用实际的例子来说就是现在我们有一堆已经标记好的西瓜，每个西瓜有自己的特征（瓜蒂大小，条纹形状，颜色深浅，....），西瓜有两个分类（这里已二分类问题为例，多分类同样适用）{好瓜，坏瓜}。
<br><br>问题：这时我们需要对这个数据集（这堆西瓜）进行建模，我们希望得到一个能用来区分西瓜好坏的模型。
<br><br>面对这个问题我们现在有两种思路：
<br><br>1、我们通过建立一个函数，计算出分类结果在特征下的条件概率。通俗的讲就是，通过一个函数计算出，在 （瓜蒂大，波浪条纹，深绿） 的条件下这个瓜是好瓜的概率是多少。
<br><br>2、我们首先建立一个概率密度模型，然后再进行推理。通俗的讲就是，我们先统计好瓜的情况下 瓜蒂大，波浪条纹，深绿 的概率分别是多少，然后对当前的西瓜进行推理。
<br><br>上面的两种方法中，第一种是判别模型的思路，第二种是生成模型的思路。
<br><br> 判别模型：有限样本==<br>判别函数 = 预测模型==<br>预测
<br> 生成模型：无穷样本==<br>概率密度模型 = 产生模型==<br>预测
<br><br>判别模型是对有限的样本进行分析然后进行预测，生成模型是先对样本进行统计，然后根据统计得出结果，在统计结果中进行预测。
<br>由此可见生成模型要求数据集要全面不然结果可能会有偏差。

### 数学推导

通过数学公式推导可以看出两种模型之间的关系：
<br>我们要有的知识储备
<br>贝叶斯公式：$p(y|x)=\frac{p(x|y)p(y)}{p(x)}$
<br>联合概率公式：$p(x,y)=\frac{p(x|y)}{p(y)}=\frac{p(y|x)}{p(x)}$
<br>首先明确我们的目标,我们希望根据特征得出预测结果。通俗的讲就是，我们希望通过西瓜的特征预测出这个西瓜是不是好瓜。
<br>用数学公式表达就是 $p(Y|X)$ 
<br>如果我们这时候直接对目标公式 $p(Y|X)$ 进行模型的分析和推导，建立的模型就是判别模型
<br>我们通过贝叶斯公式可以清楚的看到：$p(Y|X)=\frac{p(X|Y)p(Y)}{p(X)}$,我们在建模分类模型的时候是希望目标函数 $p(Y|X)$ 的结果尽可能的大，可以进行类别的区分，对$p(Y|X)$的结果的具体值是不关心的。即我希望得到的是一个西瓜是好瓜的概率大还是坏瓜的概率大，对于这个瓜是好瓜、坏瓜的概率分别是多少我们是不关心的，通过公式来看就是：
<br>$$\begin{eqnarray*}
\frac{p(Y=1|X)}{p(Y=0|X)}
&=& \frac{\frac{p(X|Y=1)p(Y=1)}{p(X)}}{\frac{p(X|Y=0)p(Y=0)}{p(X)}}\\
&=& \frac{p(X|Y=1)p(Y=1)}{p(X|Y=0)p(Y=0)}
\end{eqnarray*}$$
由此可见
$$p(Y|X) \propto p(X|Y)p(Y)$$
这样我们就可以由对 p(Y|X) 建模，改为对 p(X|Y)p(Y) 进行建模了，而且在分类任务中他们是等价的。
<br>$ p(X|Y)p(Y) = p(X,Y)$ 在这里我们可以看出生成模型是直接对联合概率进行建模的。
<br><br>对于生成模型我们同意通过边缘概率从得到：$$\sum_{i=1}^n p(X|Y_i)p(Y_i) = p(x)$$
<br>然后通过公式 $\frac {p(x,y)}{p(x)}$ 我们就从生成模型得到了判别模型，但是对于判别模型是无法得到生成模型的。 

### 优点和缺点

<br>判别模型：
<br>优点：
<br>1）仅需要有限的样本。节省计算资源，需要的样本数量也少于生成模型。
<br>2）能清晰的分辨出多类或某一类与其他类之间的差异特征，准确率往往较生成模型高。
<br>3）由于直接学习[公式]，而不需要求解类别条件概率，所以允许我们对输入进行抽象（比如降维、构造等），从而能够简化学习问题。
<br><br>缺点：
<br><br>1）不能反映训练数据本身的特性。能力有限，可以告诉你的是1还是2，但没有办法把整个场景描述出来。
<br>2）没有生成模型的优点。
<br>3）黑盒操作: 变量间的关系不清楚，不可视。
<br><br>主要应用：
<br>Image and document classification
<br>Biosequence analysis
<br>Time series prediction
<br><br>生成模型：
<br>优点：
<br>1）生成给出的是联合分布[公式]，不仅能够由联合分布计算后验分布[公式]（反之则不行），还可以给出其他信息，比如可以使用[公式]来计算边缘分布[公式]。如果一个输入样本的边缘分布[公式]很小的话，那么可以认为学习出的这个模型可能不太适合对这个样本进行分类，分类效果可能会不好，这也是所谓的异常值检测(outlier detection)。
<br>2）生成模型收敛速度比较快，即当样本数量较多时，生成模型能更快地收敛于真实模型。
<br>3）生成模型能够应付存在隐变量的情况，比如混合高斯模型就是含有隐变量的生成方法。
<br>4）研究单类问题比判别模型灵活性强
<br><br>缺点：
<br>1）联合分布能够提供更多的信息，但也需要更多的样本和更多计算，尤其是为了更准确估计类别条件分布，需要增加样本的数目，而且类别条件概率的许多信息是我们做分类用不到，因而如果我们只需要做分类任务，就浪费了计算资源。
<br>2）另外，实践中多数情况下判别模型效果更好。

### 总结

在这里我们只是通过一个简单的分类问题对生成模型和判别模型进行了简单的对比和分析，通过例子中我们可以看到：
<br>1、判别模型的思路是直接对条件概率进行建模分析的，而生成模型是对联合概率进行建模分析的。
<br>2、从生成模型可以得到判别模型，从判别模型得不到生成模型。（可以这样理解，生成模型是先对西瓜进行建模，保存了西瓜的全部特征。而判别模型是通过对不好坏西瓜的特征进行建模，只保存的是用来区分西瓜好坏的特征。所以我们可以从生成模型得到判别模型，不能从判别模型得到生成模型。）
